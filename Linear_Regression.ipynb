```python
import pandas as pd
```

```python
teams = pd.read_csv("/Users/surisettivamsikrishna/Downloads/untitled folder/Projects /Olympic data sets /teams.csv")
teams
```

```python
teams = teams.drop(columns=['height','weight'])
teams
```

```python
teams=teams.dropna()
teams
```

```python
train = teams[teams["year"]<2012]
test = teams[teams["year"] >= 2012]
```

```python
import seaborn as sns
```

```python
sns.lmplot(x='athletes',y='medals',data = teams)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
predictors = ["athletes",'prev_medals']
```

```python
reg.fit(train[predictors],train["medals"])
```

```python
Predictions = reg.predict(test[predictors])
```

```python
Predictions
```

```python
test.loc[test["predictions"] < 0, "predictions"] = 0
test
```

```python
test["predictions"] = test["predictions"].round()
```

```python
from sklearn.metrics import mean_absolute_error
error = mean_absolute_error(test["medals"],test["predictions"])
error
```
